# **Logistic Regression (One Vs All)**

### ***Importing necessary libraries and data***

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/cyberlabs/emnist_data/
train_path = '/content/drive/MyDrive/cyberlabs/emnist_data/emnist-letters-train.csv'
test_path = '/content/drive/MyDrive/cyberlabs/emnist_data/emnist-letters-test.csv'
df_train = pd.read_csv(train_path, header=None)
df_test = pd.read_csv(test_path, header=None)
```

### ***Preprocessing the Data***

```python
print('Loading Data ...\n')
X=df_train.drop([0], axis=1)
X_train=X.to_numpy().astype('float64')
X_train = X_train/255.0
m,n=X_train.shape
X_train = X_train.T
Y_train=df_train[0].to_numpy().astype('int').reshape(m, 1)
m1,n1=Y_train.shape
print(X_train,'\nShape -', m, n,'\n')
print(Y_train,'\nShape -', m1, n1)
Y_train = Y_train - 1
```

```python
print('Loading Data ...\n')
X=df_test.drop([0], axis=1)
X_test=X.to_numpy().astype('float64')
X_test = X_test/255.0
mt,nt=X_test.shape
X_test = X_test.T
Y_test=df_test[0].to_numpy().astype('int').reshape(mt, 1)
mt1,nt1=Y_test.shape
print(X_test,'\nShape -', mt, nt,'\n')
print(Y_test, '\nShape -', mt1, nt1)
Y_test = Y_test - 1
```

### ***Creating Datesets for one vs all***
Setting required output as 1 and others as 0

```python
y_train = []
for i in range(26):
  setL = []
  for j in range(88800):
      if Y_train[j] == i:
          setL.append(1)
      else:
          setL.append(0)
  y_train.append(setL)  
```

```python
y_test = []
for i in range(19):
  setM = []
  for j in range(14800):
      if Y_test[j] == i:
          setM.append(1)
      else:
          setM.append(0)
  y_test.append(setM)
```

### ***Running the algorithm***

```python
iterations = 1000
alpha = 1.2
m = 88800
costValues = []
trainedParam = []
for i in range (26):
    U = y_train[i]
    Z = np.array(U)
    weights = np.random.randn(1,784)
    bias = 0
    costList = []
    print('Training for letter', chr(i+65))
    for i in range(iterations):
       zi = np.dot(weights,X_train)+bias
       Hypothesis = 1/(1+np.exp(-zi))
       J = 1/m*(-1*(np.sum(Z*np.log(Hypothesis)+(1-Z)*np.log(1-Hypothesis))))
       costList.append(J)
       dw = 1/m*np.dot(Hypothesis-Z,X_train.T)
       db = 1/m*np.sum(Hypothesis-Z)
       weights = weights -alpha*dw
       bias = bias - alpha*db
       if((i+1)%500 == 0):
           print('Iteration number',i+1 , 'Cost', J)
       if(i==0):
           print('Iteration number',i+1 , 'Cost', J)
    costValues.append(costList)
    trainedParam.append([weights,bias])
    rng = np.arange(0, iterations)
    plt.plot(rng, costList)
    plt.show()
```

### ***Predicting the test data and calculating accuracy***

```python
Y_pred = []
a = 0
for i in range (14800):
    probabilities = []
    for j in range(19):
        weights = trainedParam[j][0]
        bias = trainedParam[j][1]
        z = np.dot(weights,X_test.T[i]) +bias
        hypothesis = 1/(1+np.exp(-z))
        probabilities.append(hypothesis)
        
    predict = probabilities.index(max(probabilities))
    predict += 1
    Y_pred.append(predict-1)
    
Y_pred = np.array(Y_pred)
Y_pred = Y_pred.T
print(Y_pred)
```

```python
count = 0
for i in range(14800):
 if (Y_pred[i] == Y_test[i]):
  count +=1 
print('Accuracy =', count/148,'%')
```
